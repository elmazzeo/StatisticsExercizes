# Exercize: Median expected significance
Derive the expression for the median expected significance and its approximation when $s \ll b$ Hint: use likelihood ratio and then use gaussian approximation of the Poisson distribution.

We consider a Poisson experiment, where the number of observed events follows a Poisson distribution with mean $\mu s+b$:
\begin{equation}
    n_{obs} \sim \text{Pois}[\mu s+b]
\end{equation}
$s$ and $b$ correspond to the number of signal and background events respectively, while $\mu$ is the signal strength.
We assume that only values of $\mu \geq 0$ make sense.

We would like to calculate the expected discovery significance $Z_{0}$:
\begin{equation}
    Z = \Phi^{-1}(1-\text{p-value})
\end{equation}
where $\Phi^{-1}$ is the inverse of the inverse of the cumulative function for the standard gaussian distribution (with mean zero and standard deviation equal to 1).

To calculate the p-value, we have to choose a test statistic: we choose the $q_{0}$ test statistic for the discovery of a positive signal.
$q_{0}$ is defined as:
\begin{equation}
    q_{0} = 
    \begin{cases}
        -2\log \lambda (0) & \hat{\mu} > 0 \\
        0 & \hat{\mu} \leq 0 \\
    \end{cases}
\end{equation}
where $\lambda(s)$ is the profile likelihood ratio:
\begin{equation}
    \lambda(\mu) = \frac{L(\mu, \hat{\hat{\theta}}(\mu))}{L(\hat{\mu}, \hat{\theta})}
\end{equation}
If we assume to know the number of background events $b$ with negligible uncertainty, we do not have nuisance parameters $\theta$ in our model.

The likelihood $L(\mu)$ is given by the Poisson distribution for the number of observed events $n$ in our counting experiment:
\begin{equation}
    L(\mu ) = \frac{(\mu s + b)^{n}}{n!}e^{-\mu s - b}
\end{equation}
and therefore:
\begin{equation}
\begin{split}
    -2\log \lambda (0) & = -2\log \frac{L(\mu )}{L(\hat{\mu} )}  \\ 
    & = -2\log \frac{\frac{(b)^{n}}{n!}e^{-b}}{\frac{(\hat{\mu} s + b)^{n}}{n!}e^{-\hat{\mu} s - b}} \\
    & = -2\log \frac{b^{n}}{(\hat{\mu} s + b)^{n}} e^{-b + \hat{\mu}s + b} \\
    & = -2\log \left [ \left(\frac{b}{\hat{\mu} s + b}\right)^{n} e^{-b + \hat{\mu}s + b} \right] \\
    & = -2\log \left [ e ^{n \log \left (\frac{b}{\hat{\mu} s + b}\right)} e^{-b + \hat{\mu}s + b}\right] \\
    & = -2\log e ^{n \log \left ( \frac{b}{\hat{\mu} s + b} \right) - b + \hat{\mu}s + b} \\
    & = -2 \left [ n \log \left ( \frac{b}{\hat{\mu} s + b} \right) - b + \hat{\mu}s + b \right ] \\
    & = -2 \left [ - n  \log \left ( \frac{\hat{\mu} s + b}{b} \right)- b + \hat{\mu}s + b \right ] \\
    & = 2 \left [ n  \log \left ( \frac{\hat{\mu} s + b}{b} \right) + b - \hat{\mu}s - b \right ] \\
    & = 2 \left [ n  \log \left ( \frac{n}{b}\right) + b - n \right] \\
\end{split}
\end{equation}
In the last step, we replaced $\hat{\mu}s + b$ with the number of observed events $n$.

The p-value is defined as:
\begin{equation}
    \text{p-value} = \int_{q_{0}}^{\infty} f(q_{0} | \mu') dq_{0}
\end{equation}
where $f(q_{0} | \mu')$ is the distribution of the $q_{0}$ test statistic under the generic signal plus background hypothesis with $\mu=\mu'$.

If we assume that the asymptotic approximation is valid, we know that
\begin{equation}
    f(q_{0} | \mu') = \left(1-\Phi \left(\frac{\mu'}{\sigma} \right) \right) \delta(q_{0}) + \frac{1}{2}\frac{1}{\sqrt{2\pi}}\frac{1}{\sqrt{q_{0}}}\exp\left[{-\frac{1}{2}\left(\sqrt{q_0} - \frac{\mu'}{\sigma}\right)^{2}}\right]
\end{equation}
therefore:
\begin{equation}
\begin{split}
    p_{0} & = \int_{q_{0}}^{\infty} f(q_{0}^{*} | \mu') dq_{0}^{*} \\
    & = \int_{q_{0}}^{\infty} \left[\left(1-\Phi \left(\frac{\mu'}{\sigma} \right) \right) \delta(q_{0}^{*}) + \frac{1}{2}\frac{1}{\sqrt{2\pi}}\frac{1}{\sqrt{q_{0}^{*}}}\exp\left[{-\frac{1}{2}\left(\sqrt{q_0^{*}} - \frac{\mu'}{\sigma}\right)^{2}}\right] \right]dq_{0^{*}} \\
    & = 0 + \int_{q_{0}}^{\infty} \frac{1}{2}\frac{1}{\sqrt{2\pi}}\frac{1}{\sqrt{q_{0}^{*}}}\exp\left[{-\frac{1}{2}\left(\sqrt{q_0^{*}} - \frac{\mu'}{\sigma}\right)^{2}}\right]dq_{0^{*}} \\
\end{split}
\end{equation}

To solve the integral, we substitute $q_{0} \rightarrow x = \sigma \sqrt{q_{0}}$. In this case:
\begin{equation}
\begin{cases}
    x = \sigma \sqrt{q_{0}} \Rightarrow  \sqrt{q_{0}} = x/\sigma \\
    \sigma^{2} q_{0} = x^{2} \Rightarrow  q_{0} = x/\sigma^{2} \\
    q_{0} \rightarrow x_{0} = \sigma \sqrt{q_{0}} \\
    dq_{0} = \frac{2x}{\sigma^{2}}dx
\end{cases}
\end{equation}
The integral becomes:
\begin{equation}
\begin{split}
    \int_{q_{0}}^{\infty}\frac{1}{2}\frac{1}{\sqrt{2\pi}}\frac{1}{\sqrt{q_{0}}}\exp\left[{-\frac{1}{2}\left(\sqrt{q_0} - \frac{\mu'}{\sigma}\right)^{2}}\right] & = \int_{x_{0}}^{\infty} \frac{1}{2}\frac{1}{\sqrt{2\pi}}\frac{\sigma}{x} \exp\left[{-\frac{1}{2}\left(\frac{x}{\sigma} - \frac{\mu'}{\sigma}\right)^{2}}\right]\frac{2x}{\sigma^{2}}dx \\
    & = \int_{x_{0}}^{\infty} \frac{1}{\sqrt{2\pi}}\frac{1}{\sigma} \exp\left[{-\frac{1}{2}\left(\frac{x - \mu'}{\sigma}\right)^{2}}\right]dx
\end{split}
\end{equation}
We end up with an integral of a gaussian distribution with mean $\mu'$ and standard deviation $\sigma$. By applying another substitution, $x\rightarrow y = (x-\mu )/\sigma$, we transform this function in a standard gaussian with mean zero and standard deviation equal to 1.

If
\begin{equation}
\begin{cases}
    x \rightarrow y = \frac{x-\mu '}{\sigma} \\
    x = \sigma y - \mu '
    x_{0} \rightarrow y_{0} = \frac{x_{0}-\mu '}{\sigma} \\
    dx = \sigma dy
\end{cases}
\end{equation}
the integral becomes:
\begin{equation}
\begin{split}
    \int_{x_{0}}^{\infty} \frac{1}{\sqrt{2\pi}}\frac{1}{\sigma} \exp\left[{-\frac{1}{2}\left(\frac{x - \mu'}{\sigma}\right)^{2}}\right]dx & = \int_{y_{0}}^{\infty} \frac{1}{\sqrt{2\pi}}\frac{1}{\sigma} e^{-y^{2}/2}\sigma dy \\
    & = \int_{y_{0}}^{\infty} \frac{1}{\sqrt{2\pi}} e^{-y^{2}/2} dy \\
    & = 1 - \int_{-\infty}^{y_{0}} \frac{1}{\sqrt{2\pi}} e^{-y^{2}/2} dy \\
    & = 1 - \Phi (y_{0}) \\
    & = 1 - \Phi \left ( \frac{x_{0}-\mu '}{\sigma}\right) \\
    & = 1 - \Phi \left ( \frac{\sigma \sqrt{q_{0}}-\mu '}{\sigma}\right) \\
    & = 1 - \Phi \left ( \sqrt{q_{0}} - \frac{\mu '}{\sigma}\right) \\
\end{split}
\end{equation}
where $\Phi$ is the cumulative function for the standard gaussian distribution.

The p-value is then:
\begin{equation}
\begin{split}
    p_{0} & = 0 + \int_{q_{0}}^{\infty} \frac{1}{2}\frac{1}{\sqrt{2\pi}}\frac{1}{\sqrt{q_{0}}}\exp\left[{-\frac{1}{2}\left(\sqrt{q_0} - \frac{\mu'}{\sigma}\right)^{2}}\right]dq_{0} \\
    & = 1 - \Phi \left ( \sqrt{q_{0}} - \frac{\mu '}{\sigma}\right)
\end{split}
\end{equation}
We can now apply the formula for the expected significance:
\begin{equation}
\begin{split}
    Z & = \Phi^{-1}(1-\text{p-value}) \\
    & = \Phi^{-1} \left(1 - 1 + \Phi \left( \sqrt{q_{0}} - \frac{\mu '}{\sigma} \right) \right) \\
    & = \Phi^{-1}\left(\Phi \left( \sqrt{q_{0}} - \frac{\mu '}{\sigma} \right)\right) \\
    & = \sqrt{q_{0}}-\frac{\mu '}{\sigma}
\end{split}
\end{equation}
In the special case where $\mu' = 0$, we obtain the formula for the expected discovery significance:
\begin{equation}
\begin{split}
    Z_{0} = \sqrt{q_{0}}
\end{split}
\end{equation}
We now consider the formula for the likelihood ratio $-2\log \lambda (0)$:
\begin{equation}
\begin{split}
    q_{0} & = - 2 \log \lambda (0) \\
    & = 2 \left[ n \log \left( \frac{n}{b} \right) + b - n \right] \\
\end{split}
\end{equation}
and we consider the case where $n = s + b$ (i.e. we obtained $\mu=1$).

The significance becomes:
\begin{split}
    Z_{0} & = \sqrt{q_{0}} \\
    & = \sqrt{2 \left[ (s+b) \log \left( \frac{s+b}{b} \right) + b - s - b \right]} \\
    & = \sqrt{2 \left[ (s+b) \log \left( 1 + s/b \right) - s \right]} \\
\end{split}



Up to now, we only have applied the asymptotic approximation. Now we assume that $ s \ll b $. If so, the logarithm can be written as:
\begin{equation}
\begin{split}
    \log (1 + s/b) \approx \frac{s}{b} - frac{1}{2}\frac{s^{2}}{b^{2}} + o\left(\frac{s^3}{b^3}\right))
\end{split}
\end{equation}
We have to include the second-order term in the expansion. Therefore, in this approximation, the formula for the expected discovery significance becomes:
\begin{equation}
\begin{split}
    Z & = \sqrt{2 \left[ (s+b) \log \left( 1 + s/b \right) - s \right]} \\
    & = \sqrt{2 \left[ (s+b) \left( \frac{s}{b} -\frac{1}{2}\frac{s^{2}}{b^{2}} + o\left(\frac{s^2}{b^2}\right) \right)- s \right]} \\
    & = \sqrt{2 \left[ \frac{s^{2}}{b} + s - \frac{1}{2}\frac{s^{3}}{b^{2}} - \frac{1}{2}\frac{s^{2}}{b} + o\left(\frac{s^2}{b^3}\right)) + o\left(\frac{s^3}{b^2}\right)) - s \right]} \\
    & = \sqrt{2 \left[ \frac{s^{2}}{b} + s - \frac{1}{2}\frac{s^{3}}{b^{2}} - \frac{1}{2}\frac{s^{2}}{b} - s + o\left(\frac{s^2}{b^2}\right))  \right]} \\
    & = \sqrt{2 \left[ \frac{1}{2}\frac{s^{2}}{b} -\frac{1}{2}\frac{s^{3}}{b^{2}} + o\left(\frac{s^2}{b^2}\right)) \right]} \\
    & = \sqrt{2 \left[ \frac{1}{2}\frac{s^{2}}{b} + o\left(\frac{s^2}{b^2}\right)) \right]} \\
    & \approx  \sqrt{\frac{s^{2}}{b}} \\
    & = \frac{s}{\sqrt{b}}
\end{split}
\end{equation}
